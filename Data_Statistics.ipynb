```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Count the occurrences of each category
category_counts = df['Value_Ebene1'].value_counts()

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")

# Formatting the chart
plt.title("Category Counts in Value_Ebene1", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

```python
# Count the occurrences of each category
category_counts = df['Value_Ebene1'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
category_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap="Set2", shadow=True)

# Formatting
plt.title("Category Proportions in Value_Ebene1", fontsize=14)
plt.ylabel("")  # Hide y-label for better aesthetics

# Show the plot
plt.show()
```
